```python
from scipy.spatial.distance import cdist, pdist
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
# Determine your k range
k_range = range(1,50)

# Fit the kmeans model for each n_clusters = k
k_means_var = [KMeans(n_clusters=k).fit(data_sns_scale) for k in k_range]
k_means_var
```

```python
# Pull out the cluster centers for each model
centroids = [X.cluster_centers_ for X in k_means_var]
centroids
```

```python
centroids[0]
```

```python
centroids[1]
```

```python
centroids[2]
```

```python
# Calculate the Euclidean distance from 
# each point to each cluster center
k_euclid = [cdist(data_sns_scale, cent, 'euclidean') for cent in centroids]
k_euclid[1]
```

```python
dist = [np.min(ke,axis=1) for ke in k_euclid]
dist[1]
```

```python
# Total within-cluster sum of squares
wcss = [sum(d**2) for d in dist]
wcss
```

```python
# The total sum of squares
tss = sum(pdist(data_sns_scale)**2)/(data_sns_scale.shape[0])
tss
```

```python
# The between-cluster sum of squares
bss = tss - wcss
bss
```

```python
# elbow curve
%matplotlib inline
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(k_range, bss/tss*100, 'g*-')
ax.set_ylim((0,100))
plt.grid(True)
plt.xlabel('Nro. clusters')
plt.ylabel('% variance explained')
plt.title('Variance Explained vs. k')
```

```python
avgWithinSumSquare = [sum(d)/data_sns_scale.shape[0] for d in dist]
avgWithinSumSquare
```

```python
# plot elbow curve
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(avgWithinSumSquare, 'b*-')
plt.grid(True)
plt.xlabel('No. of clusters')
plt.ylabel('Avg sum of squares')
tt = plt.title('Elbow Curve')
```

```python
pc_toarray=data_sns_num.values
```

```python
hpc = PCA(n_components=2).fit(pc_toarray)
hpc
```

```python
x=hpc.explained_variance_ratio_
x.sum()
```

```python
print(hpc.explained_variance_ratio_) 
```

```python
hpc = PCA(n_components=2).fit_transform(pc_toarray)
hpc
```

```python
# Determine your k range
k_range = range(1,40)

# Fit the kmeans model for each n_clusters = k
k_means_var = [KMeans(n_clusters=k).fit(hpc) for k in k_range]
k_means_var
```

```python
# Pull out the cluster centers for each model
centroids = [X.cluster_centers_ for X in k_means_var]
centroids
```

```python
# Calculate the Euclidean distance from 
# each point to each cluster center
k_euclid = [cdist(hpc, cent, 'euclidean') for cent in centroids]
k_euclid[1]
```

```python
dist = [np.min(ke,axis=1) for ke in k_euclid]
dist[1]
```

```python
avgWithinSumSquare = [sum(d)/hpc.shape[0] for d in dist]
avgWithinSumSquare
```

```python
# plot elbow curve
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(avgWithinSumSquare, 'b*-')
plt.grid(True)
plt.xlabel('No. of clusters')
plt.ylabel('Avg sum of squares')
tt = plt.title('Elbow Curve')
```
